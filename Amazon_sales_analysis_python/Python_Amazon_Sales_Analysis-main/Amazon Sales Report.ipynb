Exploratory Data Analysis


```python
df.columns
```

```python
ax = sns.countplot(x='Size',data=df)
```

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#GROUP BY
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty = df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
ay = sns.barplot(x='Size',y='Qty',data=S_Qty)
for bars in ay.containers:
    ay.bar_label(bars)
```

```python
# COURIER STATUS
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='Courier Status',hue='Status')
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
print(column_data)
```

```python
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=50)
plt.show()
```

```python
#Checking B2B data
B2B_check = df['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
plt.show()

#from pie chart we see that 99.2% buyers are retailers and 0.8% are B2B buyers
```

```python
x_data = df['Category']
y_data = df['Size']
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Category by Size')
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_states = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=50)
plt.show()
```
